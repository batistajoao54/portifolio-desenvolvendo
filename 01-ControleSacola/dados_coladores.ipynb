```python
import pandas as pd
```

```python
import tratando_os
```

```python
#carregando os dados que tem apenas as OS que ainda estão ativas
df_os = tratando_os.dados_os()
df_os
```

```python
#carregando os dados com os coladores
df_coladores = pd.read_excel("03-entrada e saida.xlsx")
df_coladores.head()
```

```python
lista_os = df_os['OS'].unique()
lista_os
```

```python
for i in lista_os:
    df_ativos = pd.read_excel('06-dados_ativos.xlsx')
    df = df_coladores[df_coladores['OS'] == i]
    df_ativos = pd.concat([df_ativos,df])
    df_ativos = df_ativos.drop_duplicates()
    df_ativos.to_excel('06-dados_ativos.xlsx', index=False)
    
```

```python
#juntando em um só arquivo os dados OS e os dados Coladores
df_final = pd.concat([df_os,df_ativos])
df_final.to_excel('04-dados-Os-Coladores-ativos.xlsx',index=False)
```
