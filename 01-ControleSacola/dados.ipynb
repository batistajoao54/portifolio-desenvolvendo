```python
import pandas as pd
```

```python
df_os = pd.read_excel('02-dados_os.xlsx')
df_os.head()
df_os.info()
df_os['OS'].unique()
```

```python
df_colador = pd.read_excel("03-entrada e saida.xlsx")
df_colador.head()
```

```python
df_recebidos = df_colador[df_colador['STATUS'] == "RECEBIDO"]
df_recebidos.head()
```

```python
df_recebidos.info()
```

```python
df_filtro_soma = df_recebidos[['OS','MARCA','TAMANHO','COR','QUANTIDADE','STATUS']].copy()
df_filtro_soma.head()
df_filtro_soma.info()
```

```python
df_soma_os = df_filtro_soma.groupby(['OS','MARCA','TAMANHO','COR','STATUS']).sum().reset_index()
df_soma_os.info()
```

```python
df_soma_os
```

```python
lista_os = df_soma_os['OS'].unique()
lista_os
```

```python
lista_os_of = df_os['OS'].unique()
lista_os_of
```

```python
#tentando comparar os dados cuidado pra nao rodar 2 vezes
for i in lista_os_of:
    df_concat = pd.read_csv('pesquisa.csv') #carregando um arquivo vazio
    df_fake = df_os[df_os['OS']== i].copy() #pesquisando cada os
    df_fake2 = df_soma_os[df_soma_os['OS'] == i].copy() #pesquisando cada os nos recebidos
    
    df_concat = pd.concat([df_fake,df_fake2,df_concat]) #juntando as informações
    df_concat= df_concat.drop_duplicates() #tentando remover as duplicatas
    df_concat.to_csv('pesquisa.csv',index=False) #salvando no mesmo diretorio csv
```

```python
df_concat
```

```python

for i in lista_os_of:
    df_ativa = pd.read_csv('ativas.csv')
    df = df_concat[df_concat['OS'] == i] #pegando cada os
    try: #criando uma condição para os que derem errado no calculo
        a = df.iloc[1,8] >= df.iloc[0,8]*0.9 #fazendo uma pequena comparação
    except: #e atribuindo o valor a False
        a = False
    
    if a == False:
        df_ativa = pd.concat([df_ativa,df]) #juntando os dados
        df_ativa = df_ativa.drop_duplicates() #tentando remover as duplicatas
        df_ativa.to_csv('ativas.csv', index=False) #salvando os dados
    
```
